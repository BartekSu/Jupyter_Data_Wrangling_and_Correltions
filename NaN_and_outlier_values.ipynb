```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('class-grades.csv',sep=',',on_bad_lines='warn')
df.head(10)
```

```python
df.isna().any()
```

```python
#NaN vlue as average
from sklearn.impute import SimpleImputer 
mean_imputer = SimpleImputer(missing_values=np.nan,strategy='mean') 
mean_imputer = mean_imputer.fit(df)
imputed_df = mean_imputer.transform(df.values) 
df = pd.DataFrame(data=imputed_df,columns=['Prefix', 'Assignment', 'Tutorial', 'Midterm', 'TakeHome', 'Final'])
df.head (10)
```

```python
df.isna().any()
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=df['Assignment'])
```

```python
filter = df['Assignment'].values>60
df_outlier_rem = df[filter]
df_outlier_rem
```
